Month-over-month revenue growth (In %)

Prompt: Create a query that displays each month's revenue and the month-over-month growth %.

```python
query_mom = """
WITH MonthlyRevenue AS (
    SELECT
        FORMAT_DATE('%Y-%m', created_at) as month,
        SUM(sale_price) as monthly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY
        1
),
MonthlyRevenueWithLag AS (
    SELECT
        month,
        monthly_revenue,
        LAG(monthly_revenue, 1) OVER (ORDER BY month) as previous_month_revenue
    FROM
        MonthlyRevenue
)
SELECT
    month,
    monthly_revenue,
    previous_month_revenue,
    (monthly_revenue - previous_month_revenue) / previous_month_revenue * 100 as month_over_month_growth
FROM
    MonthlyRevenueWithLag
ORDER BY
    month
"""
df_mom_growth = client.query(query_mom).to_dataframe()
print("Month-over-Month Revenue Growth:")
display(df_mom_growth)
```

Year-over-year revenue growth (In %)

Prompt: Create a query that displays each year's total revenue and the year-over-year growth %. Display one instance for each year rather than displaying all month, year pairs with repeating growth %.

```python
import pandas as pd

# Set pandas option to prevent scientific notation
pd.set_option('display.float_format', '{:.2f}'.format)

query_yoy = """
WITH YearlyRevenue AS (
    SELECT
        FORMAT_DATE('%Y', created_at) as year,
        SUM(sale_price) as yearly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY
        1
),
YearlyRevenueWithLag AS (
    SELECT
        year,
        yearly_revenue,
        LAG(yearly_revenue, 1) OVER (ORDER BY year) as previous_year_revenue
    FROM
        YearlyRevenue
)
SELECT
    year,
    yearly_revenue,
    previous_year_revenue,
    (yearly_revenue - previous_year_revenue) / previous_year_revenue * 100 as year_over_year_growth
FROM
    YearlyRevenueWithLag
ORDER BY
    year
"""
df_yoy_growth = client.query(query_yoy).to_dataframe()
print("Year-over-Year Revenue Growth:")
display(df_yoy_growth)
```

### Product Category: Accessories

Total accessories revenue by traffic source

Prompt: Create a query to find each marketing channel's total respective revenue for accessories.

```python
query_accessories_traffic_source = """
SELECT
    e.traffic_source,
    SUM(oi.sale_price) as total_revenue
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
JOIN
    `bigquery-public-data.thelook_ecommerce.events` e ON o.user_id = e.user_id AND DATE(o.created_at) = DATE(e.created_at) -- Attempt to join on user and date
WHERE
    p.category = 'Accessories'
GROUP BY
    1
ORDER BY
    total_revenue DESC
"""
df_accessories_traffic_source = client.query(query_accessories_traffic_source).to_dataframe()
print("Total Accessories Revenue by Traffic Source:")
display(df_accessories_traffic_source)
```

## Misleading Cross-Check

My initial query for the year-over-year growth computed the growth and displayed it, but it displayed every month, year pair with the repeated year-over-year growth % instead of just each year with the respective growth. I fixed this by fixing the query to just display the instance of each year with the respective growth %.

Prompt: Modify the query to only show one instance for each year with the respective year-over-year growth % rather than all of the month, year pairs with repeating growth %.

```python
# Original query that displayed YoY growth for each month-year pair
query_yoy_misleading = """
WITH MonthlyRevenue AS (
    SELECT
        FORMAT_DATE('%Y-%m', created_at) as month,
        SUM(sale_price) as monthly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY
        1
),
YearlyRevenue AS (
    SELECT
        FORMAT_DATE('%Y', created_at) as year,
        SUM(sale_price) as yearly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY
        1
),
YearlyRevenueWithLag AS (
    SELECT
        year,
        yearly_revenue,
        LAG(yearly_revenue, 1) OVER (ORDER BY year) as previous_year_revenue
    FROM
        YearlyRevenue
),
YearlyGrowth AS (
    SELECT
        year,
        (yearly_revenue - previous_year_revenue) / previous_year_revenue * 100 as year_over_year_growth
    FROM
        YearlyRevenueWithLag
)
SELECT
    mr.month,
    mr.monthly_revenue,
    yg.year_over_year_growth
FROM
    MonthlyRevenue mr
LEFT JOIN
    YearlyGrowth yg ON SUBSTR(mr.month, 1, 4) = yg.year
ORDER BY
    mr.month
"""
print("Misleading Output Query:")
df_yoy_misleading = client.query(query_yoy_misleading).to_dataframe()
display(df_yoy_misleading)
```

```python
# Corrected query for Year-over-Year Revenue Growth (one entry per year)
query_yoy = """
WITH YearlyRevenue AS (
    SELECT
        FORMAT_DATE('%Y', created_at) as year,
        SUM(sale_price) as yearly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items`
    GROUP BY
        1
),
YearlyRevenueWithLag AS (
    SELECT
        year,
        yearly_revenue,
        LAG(yearly_revenue, 1) OVER (ORDER BY year) as previous_year_revenue
    FROM
        YearlyRevenue
)
SELECT
    year,
    yearly_revenue,
    previous_year_revenue,
    (yearly_revenue - previous_year_revenue) / previous_year_revenue * 100 as year_over_year_growth
FROM
    YearlyRevenueWithLag
ORDER BY
    year
"""
df_yoy_growth = client.query(query_yoy).to_dataframe()
print("Corrected Output Query (Year-over-Year Revenue Growth):")
display(df_yoy_growth)
```
