```python
# Calculate 90-day Revenue Trend
# Assuming 'created_at' in 'orders' table is a timestamp or date
query_revenue_trend = """
SELECT
    DATE_TRUNC(o.created_at, DAY) as order_day,
    SUM(sale_price) as daily_revenue
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
WHERE
    o.created_at >= TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL 90 DAY)
GROUP BY
    order_day
ORDER BY
    order_day
"""
df_revenue_trend = pd.read_gbq(query_revenue_trend, project_id=project_id)
display(df_revenue_trend.tail()) # Display last few days to see recent trend
```

```python
# Calculate Repeat Purchase Rate
# This requires identifying unique customers and their order count
query_repeat_purchase = """
WITH customer_orders AS (
    SELECT
        user_id,
        COUNT(DISTINCT order_id) as order_count
    FROM
        `bigquery-public-data.thelook_ecommerce.orders`
    WHERE user_id IS NOT NULL -- Exclude orders without a user
    GROUP BY
        user_id
)
SELECT
    COUNT(DISTINCT user_id) as total_customers,
    COUNT(DISTINCT IF(order_count > 1, user_id, NULL)) as repeating_customers,
    SAFE_DIVIDE(COUNT(DISTINCT IF(order_count > 1, user_id, NULL)), COUNT(DISTINCT user_id)) as repeat_purchase_rate
FROM
    customer_orders
"""
df_repeat_purchase = pd.read_gbq(query_repeat_purchase, project_id=project_id)
display(df_repeat_purchase)
```

```python
# Calculate Average Order Value (AOV)
query_aov = """
SELECT
    SAFE_DIVIDE(SUM(sale_price), COUNT(DISTINCT order_id)) as average_order_value
FROM
    `bigquery-public-data.thelook_ecommerce.order_items`
"""
df_aov = pd.read_gbq(query_aov, project_id=project_id)
display(df_aov)
```

```python
#Calculate the Final Revenue for each month
#Prompt: Can you write code that reports the final revenue for each month
query_monthly_revenue = """
SELECT
    DATE_TRUNC(o.created_at, MONTH) as order_month,
    SUM(oi.sale_price) as monthly_revenue
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
GROUP BY
    order_month
ORDER BY
    order_month
"""
df_monthly_revenue = pd.read_gbq(query_monthly_revenue, project_id=project_id)
display(df_monthly_revenue)
```

```python
# Unique Customers placing orders each month
# Can you calculate the number of unique customers that are placing orders each month
query_unique_customers_month = """
SELECT
    DATE_TRUNC(created_at, MONTH) as order_month,
    COUNT(DISTINCT user_id) as unique_customers
FROM
    `bigquery-public-data.thelook_ecommerce.orders`
WHERE user_id IS NOT NULL
GROUP BY
    order_month
ORDER BY
    order_month
"""
df_unique_customers_month = pd.read_gbq(query_unique_customers_month, project_id=project_id)
display(df_unique_customers_month)
```

```python
# Total orders per month
# Prompt: Can you calculate the total number of orders each month
query_monthly_orders = """
SELECT
    DATE_TRUNC(created_at, MONTH) as order_month,
    COUNT(order_id) as total_orders
FROM
    `bigquery-public-data.thelook_ecommerce.orders`
GROUP BY
    order_month
ORDER BY
    order_month
"""
df_monthly_orders = pd.read_gbq(query_monthly_orders, project_id=project_id)
display(df_monthly_orders)
```

```python
# Avg cost of each order each month
# Can you calculate the average cost of an order each month
query_average_order_cost_month = """
SELECT
    DATE_TRUNC(o.created_at, MONTH) as order_month,
    AVG(ii.cost) as average_order_cost
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
JOIN
    `bigquery-public-data.thelook_ecommerce.inventory_items` ii ON oi.inventory_item_id = ii.id
GROUP BY
    order_month
ORDER BY
    order_month
"""
df_average_order_cost_month = pd.read_gbq(query_average_order_cost_month, project_id=project_id)
display(df_average_order_cost_month)
```

```python
# Total revenue per month
# Prompt: can you write code using both CTEs and Window Functions to calculate the final revenue per month
query_monthly_revenue_cte_window = """
WITH MonthlyRevenue AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        SUM(oi.sale_price) as monthly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
    GROUP BY
        order_month
)
SELECT
    order_month,
    monthly_revenue,
    SUM(monthly_revenue) OVER (ORDER BY order_month) as cumulative_revenue
FROM
    MonthlyRevenue
ORDER BY
    order_month
"""
df_monthly_revenue_cte_window = pd.read_gbq(query_monthly_revenue_cte_window, project_id=project_id)
display(df_monthly_revenue_cte_window)
```

Since the overall number of orders is rising every month, it is hard to see what that means for tops and tee sales on an individual basis. This means I need to look at this statistic as a percent of total sales.

```python
# Percentage of Total Sales for 'Tops & Tees' Category by Month
# Prompt: can you find the %age that the tops and tees category makes up of total sales on a monthly basis?
query_tops_tees_percentage = """
WITH MonthlyTotalRevenue AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        SUM(oi.sale_price) as total_monthly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items` oi
    JOIN
        `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
    GROUP BY
        order_month
),
MonthlyTopsTeesRevenue AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        SUM(oi.sale_price) as tops_tees_monthly_revenue
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items` oi
    JOIN
        `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    WHERE
        p.category = 'Tops & Tees'
    GROUP BY
        order_month
)
SELECT
    t.order_month,
    t.total_monthly_revenue,
    tt.tops_tees_monthly_revenue,
    SAFE_DIVIDE(tt.tops_tees_monthly_revenue, t.total_monthly_revenue) * 100 as tops_tees_percentage_of_total_sales
FROM
    MonthlyTotalRevenue t
JOIN
    MonthlyTopsTeesRevenue tt ON t.order_month = tt.order_month
ORDER BY
    t.order_month
"""
df_tops_tees_percentage = pd.read_gbq(query_tops_tees_percentage, project_id=project_id)
display(df_tops_tees_percentage)
```

```python
# Percentage of Total Orders for 'Tops & Tees' by Month and Location, Top/Bottom 10 Locations
# Prompt: can you find out how many the %age of total orders were for tps & tees, by month and location, and list the top 10 places and the bottom 10 places?
query_tops_tees_location_percentage = """
WITH MonthlyLocationTotalOrders AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        u.state,
        COUNT(DISTINCT o.order_id) as total_monthly_distinct_orders_location
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.users` u ON o.user_id = u.id
    WHERE u.state IS NOT NULL
    GROUP BY
        order_month,
        u.state
),
MonthlyLocationTopsTeesOrders AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        u.state,
        COUNT(DISTINCT o.order_id) as tops_tees_monthly_distinct_orders_location
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    JOIN
        `bigquery-public-data.thelook_ecommerce.users` u ON o.user_id = u.id
    WHERE p.category = 'Tops & Tees' AND u.state IS NOT NULL
    GROUP BY
        order_month,
        u.state
),
MonthlyLocationTopsTeesPercentage AS (
    SELECT
        mlto.order_month,
        mlto.state,
        mlto.total_monthly_distinct_orders_location,
        COALESCE(mltto.tops_tees_monthly_distinct_orders_location, 0) as tops_tees_monthly_distinct_orders_location,
        SAFE_DIVIDE(COALESCE(mltto.tops_tees_monthly_distinct_orders_location, 0), mlto.total_monthly_distinct_orders_location) * 100 as tops_tees_percentage
    FROM
        MonthlyLocationTotalOrders mlto
    LEFT JOIN
        MonthlyLocationTopsTeesOrders mltto ON mlto.order_month = mltto.order_month AND mlto.state = mltto.state
)
SELECT
    order_month,
    state,
    tops_tees_percentage
FROM
    MonthlyLocationTopsTeesPercentage
WHERE total_monthly_distinct_orders_location > 0 -- Avoid division by zero and cases with no orders
ORDER BY
    tops_tees_percentage DESC
LIMIT 10
"""
df_tops_10_locations = pd.read_gbq(query_tops_tees_location_percentage, project_id=project_id)
display(df_tops_10_locations)

query_bottom_tops_tees_location_percentage = """
WITH MonthlyLocationTotalOrders AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        u.state,
        COUNT(DISTINCT o.order_id) as total_monthly_distinct_orders_location
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.users` u ON o.user_id = u.id
    WHERE u.state IS NOT NULL
    GROUP BY
        order_month,
        u.state
),
MonthlyLocationTopsTeesOrders AS (
    SELECT
        DATE_TRUNC(o.created_at, MONTH) as order_month,
        u.state,
        COUNT(DISTINCT o.order_id) as tops_tees_monthly_distinct_orders_location
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    JOIN
        `bigquery-public-data.thelook_ecommerce.users` u ON o.user_id = u.id
    WHERE p.category = 'Tops & Tees' AND u.state IS NOT NULL
    GROUP BY
        order_month,
        u.state
),
MonthlyLocationTopsTeesPercentage AS (
    SELECT
        mlto.order_month,
        mlto.state,
        mlto.total_monthly_distinct_orders_location,
        COALESCE(mltto.tops_tees_monthly_distinct_orders_location, 0) as tops_tees_monthly_distinct_orders_location,
        SAFE_DIVIDE(COALESCE(mltto.tops_tees_monthly_distinct_orders_location, 0), mlto.total_monthly_distinct_orders_location) * 100 as tops_tees_percentage
    FROM
        MonthlyLocationTotalOrders mlto
    LEFT JOIN
        MonthlyLocationTopsTeesOrders mltto ON mlto.order_month = mltto.order_month AND mlto.state = mltto.state
)
SELECT
    order_month,
    state,
    tops_tees_percentage
FROM
    MonthlyLocationTopsTeesPercentage
WHERE total_monthly_distinct_orders_location > 0 -- Avoid division by zero and cases with no orders
ORDER BY
    tops_tees_percentage ASC
LIMIT 10
"""
df_bottom_10_locations = pd.read_gbq(query_bottom_tops_tees_location_percentage, project_id=project_id)
display(df_bottom_10_locations)
```

```python
# Distribution of Browsers for Tops & Tees Sales
# Prompt: using the list you just generated, can you report to me what device the majority of these places that have over a 50% tops_tees_percentage use to place their order.
# This query is adjusted to show the distribution of all browsers for Tops & Tees sales.
# Prompt: can you adjust it to show the distribution of all browsers
query_tops_tees_browser_distribution = """
SELECT
    e.browser,
    COUNT(DISTINCT o.order_id) as tops_tees_orders_by_device
FROM
    `bigquery-public-data.thelook_ecommerce.orders` o
JOIN
    `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
JOIN
    `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
JOIN
    `bigquery-public-data.thelook_ecommerce.events` e ON o.user_id = e.user_id -- Joined on user_id
WHERE
    p.category = 'Tops & Tees'
    AND e.browser IS NOT NULL
    AND e.event_type = 'purchase' -- Filter for purchase events
GROUP BY
    e.browser
ORDER BY
    tops_tees_orders_by_device DESC
"""
df_tops_tees_browser_distribution = pd.read_gbq(query_tops_tees_browser_distribution, project_id=project_id)
display(df_tops_tees_browser_distribution)
```

```python
# 3rd Most Popular Browser for Tops & Tees Sales
# Prompt: can you list the 3rd most popular browser
query_tops_tees_3rd_most_popular_browser = """
WITH TopsTeesBrowserRank AS (
    SELECT
        e.browser,
        COUNT(DISTINCT o.order_id) as tops_tees_orders_by_device,
        RANK() OVER (ORDER BY COUNT(DISTINCT o.order_id) DESC) as rank_num
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    JOIN
        `bigquery-public-data.thelook_ecommerce.events` e ON o.user_id = e.user_id -- Joined on user_id
    WHERE
        p.category = 'Tops & Tees'
        AND e.browser IS NOT NULL
        AND e.event_type = 'purchase' -- Filter for purchase events
    GROUP BY
        e.browser
)
SELECT
    browser,
    tops_tees_orders_by_device
FROM
    TopsTeesBrowserRank
WHERE rank_num = 3
"""
df_tops_tees_3rd_most_popular_browser = pd.read_gbq(query_tops_tees_3rd_most_popular_browser, project_id=project_id)
display(df_tops_tees_3rd_most_popular_browser)
```

```python
# Percentage Distribution of Traffic Source for Tops & Tees Sales
# Prompt: can you find out what the percentage distribution source of traffic is for tops and tees
query_tops_tees_traffic_source_percentage = """
WITH TopsTeesTotalOrders AS (
    SELECT
        COUNT(DISTINCT o.order_id) as total_tops_tees_orders
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    WHERE
        p.category = 'Tops & Tees'
),
TopsTeesOrdersByTrafficSource AS (
    SELECT
        e.traffic_source,
        COUNT(DISTINCT o.order_id) as tops_tees_orders_by_source
    FROM
        `bigquery-public-data.thelook_ecommerce.orders` o
    JOIN
        `bigquery-public-data.thelook_ecommerce.order_items` oi ON o.order_id = oi.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
    JOIN
        `bigquery-public-data.thelook_ecommerce.events` e ON o.user_id = e.user_id -- Joined on user_id
    WHERE
        p.category = 'Tops & Tees'
        AND e.traffic_source IS NOT NULL
        AND e.event_type = 'purchase' -- Filter for purchase events
    GROUP BY
        e.traffic_source
)
SELECT
    t.traffic_source,
    t.tops_tees_orders_by_source,
    SAFE_DIVIDE(t.tops_tees_orders_by_source, tt.total_tops_tees_orders) * 100 as percentage_of_tops_tees_orders
FROM
    TopsTeesOrdersByTrafficSource t
JOIN
    TopsTeesTotalOrders tt ON 1=1 -- Join to get the total count
ORDER BY
    percentage_of_tops_tees_orders DESC
"""
df_tops_tees_traffic_source_percentage = pd.read_gbq(query_tops_tees_traffic_source_percentage, project_id=project_id)
display(df_tops_tees_traffic_source_percentage)
```

```python
# Interactive Plotly Chart: 30-Day Revenue Scorecard
# Prompt: can you help me generate an interactive plotly chart with the scorecard (revenue for the last 30 days)

import plotly.graph_objects as go
import pandas as pd

# Calculate 30-day Revenue
query_30day_revenue = """
SELECT
    SUM(oi.sale_price) as total_revenue_last_30_days
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
WHERE
    o.created_at >= TIMESTAMP_SUB(CURRENT_TIMESTAMP(), INTERVAL 30 DAY)
"""
df_30day_revenue = pd.read_gbq(query_30day_revenue, project_id=project_id)

# Get the total revenue value
total_revenue = df_30day_revenue['total_revenue_last_30_days'].iloc[0]

# Create a Plotly Indicator gauge chart for the scorecard
fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = total_revenue,
    title = {'text': "Total Revenue Last 30 Days"},
    gauge = {'axis': {'visible': False}} # Hide the gauge axis for a simple scorecard
))

fig.update_layout(
    margin=dict(l=20, r=20, t=40, b=20),
    height=200,
    title_x=0.5 # Center the title
)

fig.show()
```

```python
# Plotly Pie Chart: Sales Percentage by Region (State) with Order Ratio in Hover
# Prompt: can you create a plotly chart that has a pie chart of sales% by region?
# Prompt: would it be possible to include the number of total orders made in this location / total orders made in the hover menu?

import plotly.express as px
import pandas as pd

# Calculate Sales and Total Orders by Region (State) and Overall Total Orders
query_sales_orders_by_region = """
WITH StateSalesAndOrders AS (
    SELECT
        u.state,
        SUM(oi.sale_price) as total_sales,
        COUNT(DISTINCT o.order_id) as total_orders_state
    FROM
        `bigquery-public-data.thelook_ecommerce.order_items` oi
    JOIN
        `bigquery-public-data.thelook_ecommerce.orders` o ON oi.order_id = o.order_id
    JOIN
        `bigquery-public-data.thelook_ecommerce.users` u ON o.user_id = u.id
    WHERE u.state IS NOT NULL
    GROUP BY
        u.state
),
TotalOrdersOverall AS (
    SELECT
        COUNT(DISTINCT order_id) as total_orders_overall
    FROM
        `bigquery-public-data.thelook_ecommerce.orders`
)
SELECT
    s.state,
    s.total_sales,
    s.total_orders_state,
    t.total_orders_overall
FROM
    StateSalesAndOrders s
CROSS JOIN
    TotalOrdersOverall t
ORDER BY
    s.total_sales DESC
"""
print(f"Using project_id: {project_id}")
df_sales_orders_by_region = pd.read_gbq(query_sales_orders_by_region, project_id=project_id)

# Calculate the ratio of total orders in this location to total orders overall
df_sales_orders_by_region['order_ratio'] = df_sales_orders_by_region['total_orders_state'] / df_sales_orders_by_region['total_orders_overall']

# Create a Plotly pie chart
# Include 'total_orders_state' and 'order_ratio' in hover data
fig = px.pie(df_sales_orders_by_region, values='total_sales', names='state', title='Sales Percentage by State',
             hover_data=['total_sales', 'total_orders_state', 'order_ratio'])

# Update the textinfo to show labels on hover
fig.update_traces(textinfo='none')
fig.show()
```

```python
# Plotly Bar Chart: Top 5 Product Categories by Sales
# Prompt: what about creating a bar chart that lists the top 5 products/categories?

import plotly.express as px
import pandas as pd

# Calculate Sales by Product Category
query_sales_by_category = """
SELECT
    p.category,
    SUM(oi.sale_price) as total_sales
FROM
    `bigquery-public-data.thelook_ecommerce.order_items` oi
JOIN
    `bigquery-public-data.thelook_ecommerce.products` p ON oi.product_id = p.id
GROUP BY
    p.category
ORDER BY
    total_sales DESC
LIMIT 5
"""
df_sales_by_category = pd.read_gbq(query_sales_by_category, project_id=project_id)

# Create a Plotly bar chart
fig = px.bar(df_sales_by_category, x='category', y='total_sales', title='Top 5 Product Categories by Sales')
fig.show()
```
